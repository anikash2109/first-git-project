```python
class BankAccount:

    def __init__(self, acc_holder, balance):
        self.acc_holder=acc_holder
        self.balance = balance

    def deposit(self, amount):
        if amount<=100:
            print("Cannot Deposit such small amount")
        else:
            self.balance+=amount
            print(f" Amount added: {amount}")
            print(f" Updated Balance: {self.balance}")
           
    
    def withdrawn(self,amount):
        if amount>self.balance:
            print("Cannot withdrawn due to insufficient balance")
        else:
            self.balance-=amount
            print(F" Successfully withdrawn : {self.balance}")



ani = BankAccount("Aniket", 1000)

ani.deposit(500)
```
